```python
import pandas as pd
import os

folder_path = 'data/'  # data 폴더의 경로를 설정합니다.
excel_files = [f for f in os.listdir(folder_path) if f.endswith('.xlsx')]  # 엑셀 파일만 선택합니다.

dfs = {}  # 데이터프레임을 저장할 빈 딕셔너리를 생성합니다.

for file_name in excel_files:
    df_name = os.path.splitext(file_name)[0]  # 파일 이름에서 확장자를 제거하여 데이터프레임의 이름으로 사용합니다.
    df = pd.read_excel(os.path.join(folder_path, file_name), index_col=0)  # 엑셀 파일을 데이터프레임으로 불러옵니다.
    dfs[df_name] = df  # 딕셔너리에 데이터프레임을 추가합니다.

```

```python
from hanspell import spell_checker
from pykospacing import Spacing
import re

#Review-Title
# 띄어쓰기 조정을 위한 함수 초기화
spacing = Spacing()

for df_name, df in dfs.items():
    print(df_name)
    if 'Review' in df.columns:
        for idx, row in df.iterrows():
            review = str(row['Title'])

            # 특수 문자 제거
            review = re.sub(r'[^가-힣0-9a-zA-Z\s]', '', review)

            # 리뷰 길이가 지나치게 길 경우 분할
            max_length = 200  # 적절한 길이 설정
            review_parts = [review[i:i+max_length] for i in range(0, len(review), max_length)]

            corrected_parts = []
            for part in review_parts:
                # 오타 수정
                try:
                    spelled_review = spell_checker.check(part)
                    corrected_review = spelled_review.checked
                    corrected_parts.append(corrected_review)
                except Exception as e:
                    print(f"Error processing review part: {part[:30]}... Error: {e}")

            # 모든 종류의 공백 문자 제거 및 띄어쓰기 조정
            corrected_text = ' '.join(corrected_parts)
            corrected_text = spacing(corrected_text)

            # 데이터프레임 업데이트
            df.at[idx, 'Title'] = corrected_text
```

```python
from konlpy.tag import Okt

# Okt 형태소 분석기 초기화
okt = Okt()

# dfs에 있는 모든 데이터프레임에 대해 형태소 분석 수행
for df_name, df in dfs.items():
    print(f"Processing {df_name}...")
    if 'Review' in df.columns:
        # 형태소 분석 결과를 저장할 리스트 초기화
        df['Morphs'] = None  # 새로운 열을 생성

        for idx, row in df.iterrows():
            review = str(row['Review'])
            # 형태소 분석 및 조사 제거
            morphs = okt.morphs(review, stem=True)
            # 결과를 해당 행에 저장
            df.at[idx, 'Morphs'] = morphs
```

```python
# 불용어 목록 정의
stopwords = ['의', '가', '이', '은', '들', '는', '좀', '잘', '걍', '과', '도', '를', '으로', '자', '에', '와', '한', '하다']

# dfs에 있는 모든 데이터프레임에 대해 불용어 제거 및 문자열 형태로 저장
for df_name, df in dfs.items():
    print(f"Processing {df_name}...")
    if 'Morphs' in df.columns:  # 형태소 분석 결과가 있는지 확인
        # 불용어 제거 결과를 저장할 리스트 초기화
        df['Morphs_Cleaned'] = None  # 새로운 열을 생성

        for idx, row in df.iterrows():
            morphs = row['Morphs']
            # 불용어 제거
            morphs_cleaned = [word for word in morphs if word not in stopwords]
            # 결과를 문자열로 변환하여 해당 행에 저장
            df.at[idx, 'Morphs_Cleaned'] = " ".join(morphs_cleaned)
```

## 감성분석

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# 감성 분석기를 초기화합니다.
senti_analyzer = SentimentIntensityAnalyzer()

# 각 데이터프레임에 대해 감성 분석 수행
for df_name, df in dfs.items():
    print(f"Processing {df_name}...")
    if 'Review' in df.columns:  # 이미 불용어가 처리된 텍스트 데이터가 있는 열을 확인합니다.
        df['sent'] = None  # 새로운 열을 생성하여 감성 점수를 저장할 준비를 합니다.

        for idx, row in df.iterrows():
            text_cleaned = row['Review']
#             text_cleaned = row['Title']
            
            # 감성 분석 수행
            senti_scores = senti_analyzer.polarity_scores(text_cleaned[:30])
            
            # 레코드 값이 '1'인 경우 'compound' 값을 -1로 설정
            if text_cleaned == '1':
                senti_scores['compound'] = -1
                senti_scores['pos'] = -1
                senti_scores['neg'] = -1
                senti_scores['neu'] = -1
            # 감성 점수를 해당 행에 저장 (예: 'compound' 점수)
            df.at[idx, 'sent_cp'] = senti_scores['compound']
            df.at[idx, 'sent_ps'] = senti_scores['pos']
            df.at[idx, 'sent_ng'] = senti_scores['neg']
            df.at[idx, 'sent_nu'] = senti_scores['neu']

```

```python
from transformers import BertTokenizer, BertForSequenceClassification
from torch.utils.data import DataLoader, TensorDataset
import torch

# KoBERT 모델과 토크나이저 로드
model = BertForSequenceClassification.from_pretrained("monologg/kobert")
tokenizer = BertTokenizer.from_pretrained("monologg/kobert")

# 각 데이터프레임에 대해 감성 분석 수행
for df_name, df in dfs.items():
    print(f"Processing {df_name}...")
    if 'Review' in df.columns:  # 이미 불용어가 처리된 텍스트 데이터가 있는 열을 확인합니다.
        df['sent'] = None  # 새로운 열을 생성하여 감성 점수를 저장할 준비를 합니다.

        for idx, row in df.iterrows():
            text_cleaned = row['Review'][:30]  # 처음 30자까지만 사용
            
            # 문장을 KoBERT 모델 입력 형식으로 변환
            inputs = tokenizer(text_cleaned, return_tensors='pt', truncation=True, padding=True, max_length=64)
            
            # 모델로부터 감성 점수 계산
            outputs = model(**inputs)
            logits = outputs.logits
            probabilities = torch.softmax(logits, dim=1)
            
            # 긍정 및 부정 감성 점수 추출
            pos_score = probabilities[0][1].item()
            neg_score = probabilities[0][0].item()
            
            # 레코드 값이 '1'인 경우 'compound' 값을 -1로 설정
            if text_cleaned == '1':
                pos_score = -1
                neg_score = -1
            
            # 감성 점수를 해당 행에 저장
            df.at[idx, 'sent_ps'] = pos_score
            df.at[idx, 'sent_ng'] = neg_score
```

### 풀무원 부정

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np

# '동원'을 f-string으로 변수로 정의
company_name = '풀무원'

# 다섯 개의 데이터프레임에서 Review 칼럼을 추출하여 하나의 리스트로 합칩니다.
review_texts = []
for df in [f'{company_name}_best_star1', f'{company_name}_best_star2']:
    review_texts.extend(dfs[df]['Morphs_Cleaned'].tolist())

# NaN 값을 빈 문자열로 대체
review_texts = [str(text) for text in review_texts]
review_texts = [text for text in review_texts if text != '1']


# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(review_texts)

# K-Means 군집화
n_clusters = 3  # 군집의 개수
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_


def get_top_features_cluster(tfidf_matrix, prediction, n_feats, features):
    dfs = []
    for cluster in range(n_clusters):
        id_temp = np.where(prediction == cluster)[0]  # 해당 군집에 속하는 데이터 인덱스
        x_means = np.asarray(np.mean(tfidf_matrix[id_temp], axis=0)).flatten()  # 해당 군집의 평균 TF-IDF 점수, 배열로 변환
        
        sorted_means = np.argsort(x_means)[::-1]  # 역순으로 정렬
        best_features = [(features[int(i)], x_means[int(i)]) for i in sorted_means[:n_feats]]

        df = pd.DataFrame(best_features, columns=['단어', '점수'])
        dfs.append(df)
    return dfs

# TF-IDF 벡터라이저에서 특성 이름들 추출
feature_names = vectorizer.get_feature_names_out()

clusters = kmeans.labels_

# 이후에 함수 호출
top_features_per_cluster = get_top_features_cluster(X, clusters, 20, feature_names)


# 결과 출력
for i, top_words_df in enumerate(top_features_per_cluster):
    print(f"{company_name} 부정 군집 {i + 1} 주요 단어:")
    print(top_words_df)
    print()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import koreanize_matplotlib

# PCA를 이용한 차원 축소
pca = PCA(n_components=2)  # 2차원으로 축소
X_reduced = pca.fit_transform(X.toarray())

# 클러스터링 결과 시각화
plt.figure(figsize=(10, 10))
for i in range(n_clusters):
    plt.scatter(X_reduced[labels == i, 0], X_reduced[labels == i, 1], label=f"Cluster {i + 1}")
plt.title(f"{company_name} 부정 TF-IDF K-Means Clustering")
plt.legend()
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

# 코사인 유사도 계산
cosine_sim = cosine_similarity(X.toarray())

# 계층적 클러스터링
linked = linkage(cosine_sim, method='ward')

# 덴드로그램 그리기
plt.figure(figsize=(10, 7))
dendrogram(linked, 
           orientation='top', 
           distance_sort='descending', 
           show_leaf_counts=True, 
           color_threshold=4,  # 클러스터 색상 변경 임계값 설정
           truncate_mode='lastp',  # 덴드로그램을 자를 방식 설정
           p = 50)  # 표시할 최대 클러스터 개수 설정
plt.title(f"{company_name} 부정 Hierarchical Clustering Dendrogram")
plt.axhline(y=4, color='r', linestyle='--')  # 클러스터링 높이 표시
plt.show()

# 높이 5에서 클러스터 형성
clusters = fcluster(linked, 4, criterion='distance')

def get_top_features_per_cluster(tfidf_matrix, clusters, feature_names, top_n):
    cluster_top_features = {}
    for cluster_id in np.unique(clusters):
        cluster_indices = np.where(clusters == cluster_id)[0]
        # 유효한 인덱스만 사용
        cluster_indices = cluster_indices[cluster_indices < tfidf_matrix.shape[0]]
        if len(cluster_indices) > 0:
            mean_tfidf = np.mean(tfidf_matrix[cluster_indices].toarray(), axis=0)
            top_feature_indices = mean_tfidf.argsort()[::-1][:top_n]
            cluster_top_features[cluster_id] = [feature_names[i] for i in top_feature_indices]
    return cluster_top_features

# 각 클러스터의 상위 단어 추출
top_features_per_cluster = get_top_features_per_cluster(X, clusters, feature_names, 5)

# 결과 출력
for cluster_id, top_features in top_features_per_cluster.items():
    print(f"Cluster {cluster_id}: {', '.join(top_features)}")


```
