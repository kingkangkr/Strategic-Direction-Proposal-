```python
#1-1
!pip install Afinn
import pandas as pd
import nltk
from afinn import Afinn
nltk.download ('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import RegexpTokenizer
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount ('/content/gdrive')
# Specify the file name and save path
file_name = "/content/gdrive/MyDrive/IMDB Dataset.csv"
review = pd.read_csv ( file_name , engine="python")

from afinn import Afinn
afinn = Afinn()
pos_review = review['review'][1] # Print only one positive sentence
neg_review = review['review'][3] # print only one negative sentence


file_name1 = "/content/gdrive/MyDrive/NRC-Emotion-Lexicon-Wordlevel-v0.92.txt"
NRC = pd.read_csv(file_name1, engine="python", header=None, sep="\t")
train_y = review['sentiment'][0:700]  # 70% train dataset
test_y = review['sentiment'][700:1000] # 30% test dataset
# AFINN sentiment score (AFINN: -5 ~ 5)
afinn = Afinn()
n = 1000
score_list = []
for row in review['review'][0:n]:
    afscore = afinn.score(row)
    score_list.append(afscore)
train_x = score_list[0:700]
test_x = score_list[700:1000]
###### LogisticRegression #########################################
import numpy as np
from sklearn.linear_model import LogisticRegression, SGDClassifier
model = LogisticRegression()
# Independent & dependent variable should be array type
model.fit(np.array(train_x).reshape(-1,1), np.array(train_y))  # If there is only one independent variable, use reshqpe(-1,1)
# model estimation
print(model.score(np.array(train_x).reshape(-1,1), np.array(train_y)))
#(1) prediction
print(model.score(np.array(test_x).reshape(-1,1), np.array(test_y)))  # accuracy calculation
####################################
# Initialize an empty list to store sentiment scores
```

```python

#question
n = 1000
score_list = []
tokenizer = RegexpTokenizer('[\w]+')
stop_words = set(stopwords.words('english'))
for row in review['review'][0:n]:
  raw = row.lower()
  tokens = tokenizer.tokenize(raw)
  stopped_tokens = [i for i in tokens if not i in stop_words] # remove stopwords
  match_words = [x for x in stopped_tokens if x in list(NRC[0])] # match w/ lexicon

  emotion=[]
  for i in match_words:
    temp = list(NRC.iloc[np.where(NRC[0] == i)[0],1])  # emotion list
    for j in temp:
      emotion.append(j)
  sentiment_result1 = pd.Series(emotion).value_counts()
  try:
    neg = sentiment_result1.negative  # negative word count
  except:
    neg = 0
  try:
    pos = sentiment_result1.positive # positive word count
  except:
    pos = 0
  score = pos-neg   #
  score_list.append(score)
```

```python
#1-2
# Add the sentiment scores to the DataFrame
reviewx['sentiment_score'] = score_list[:n]

# Now, the sentiment scores should match the length of the subset of your data

# Split the data into train and test sets
train_x_nrc = reviewx['sentiment_score'][:700].values.reshape(-1, 1)
test_x_nrc = reviewx['sentiment_score'][700:1000].values.reshape(-1, 1)

# Logistic Regression with NRC sentiment score
model_nrc = LogisticRegression()
model_nrc.fit(train_x_nrc, np.array(train_y))

# Model evaluation for NRC sentiment score
print("NRC Model Accuracy - Train:", model_nrc.score(train_x_nrc, np.array(train_y)))
print("NRC Model Accuracy - Test:", model_nrc.score(test_x_nrc, np.array(test_y)))
```
